# Thuật Toán Bayes Classification Methods

```python
import pandas as pd

data = [
    ['sunny','hot','high','weak','no'],
    ['sunny','hot','high','strong','no'],
    ['overcast','hot','high','weak','yes'],
    ['rainy','mild','high','weak','yes'],
    ['rainy','cool','normal','weak','yes'],
    ['rainy','cool','normal','strong','no'],
    ['overcast','cool','normal','strong','yes'],
    ['sunny','mild','high','weak','no'],
    ['sunny','cool','normal','weak','yes'],
    ['rainy','mild','normal','weak','yes'],
    ['sunny','mild','normal','strong','yes'],
    ['overcast','mild','high','strong','yes'],
    ['overcast','hot','normal','weak','yes'],
    ['rainy','mild','high','strong','no']
]

df = pd.DataFrame(
    data,
    columns=[
        'outlook',
        'temperature',
        'humidity',
        'wind',
        'play'
    ]
)

outlook = input("Outlook: ")
temperature = input("Temperature: ")
humidity = input("Humidity: ")
wind = input("Wind: ")

yes_data = df[df['play'] == 'yes']

p_yes = len(yes_data) / len(df)

p_yes *= len(yes_data[yes_data['outlook'] == outlook]) / len(yes_data)
p_yes *= len(yes_data[yes_data['temperature'] == temperature]) / len(yes_data)
p_yes *= len(yes_data[yes_data['humidity'] == humidity]) / len(yes_data)
p_yes *= len(yes_data[yes_data['wind'] == wind]) / len(yes_data)

no_data = df[df['play'] == 'no']

p_no = len(no_data) / len(df)

p_no *= len(no_data[no_data['outlook'] == outlook]) / len(no_data)
p_no *= len(no_data[no_data['temperature'] == temperature]) / len(no_data)
p_no *= len(no_data[no_data['humidity'] == humidity]) / len(no_data)
p_no *= len(no_data[no_data['wind'] == wind]) / len(no_data)

print("\nKết quả:")
print("P(yes) =", p_yes)
print("P(no) =", p_no)

if p_yes > p_no:
    print("Dự đoán: YES")
else:
    print("Dự đoán: NO")
```
