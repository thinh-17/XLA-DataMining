# Với data như sau:
| Điểm | X   | Y  |
|------|----:|---:|
| P1   | 1   | 1  |
| P2   | 1   | 2  |
| P3   | 2   | 1  |
| P4   | 2   | 2  |
| P5   | 7   | 7  |
| P6   | 7   | 8  |
| P7   | 8   | 7  |
| P8   | 8   | 8  |
| P9   | 13  | 1  |
| P10  | 13  | 2  |
| P11  | 14  | 1  |
| P12  | 14  | 2  |

```python
import math

data = [
    [1, 1],
    [1, 2],
    [2, 1],
    [2, 2],

    [7, 7],
    [7, 8],
    [8, 7],
    [8, 8],

    [13, 1],
    [13, 2],
    [14, 1],
    [14, 2]
]

k = 3

# Chọn tâm cụm ban đầu
centroids = [
    data[0].copy(),   # [1, 1]
    data[4].copy(),   # [7, 7]
    data[8].copy()    # [13, 1]
]


def euclidean_distance(point, centroid):
    total = 0

    for i in range(len(point)):
        total += (point[i] - centroid[i]) ** 2

    return math.sqrt(total)


def calculate_centroid(cluster):
    if len(cluster) == 0:
        return None

    new_centroid = []

    for column in range(len(cluster[0])):
        total = 0

        for point in cluster:
            total += point[column]

        average = total / len(cluster)
        new_centroid.append(average)

    return new_centroid


def centroids_equal(old_centroids, new_centroids):
    epsilon = 0.000001

    for i in range(len(old_centroids)):
        for j in range(len(old_centroids[i])):
            if abs(old_centroids[i][j] - new_centroids[i][j]) > epsilon:
                return False

    return True


max_iterations = 100

for iteration in range(1, max_iterations + 1):
    print("\n" + "=" * 60)
    print(f"VÒNG LẶP {iteration}")
    print("=" * 60)

    print("Tâm cụm hiện tại:")

    for i, centroid in enumerate(centroids, start=1):
        print(f"C{i} = {centroid}")

    # Tạo 3 cụm rỗng
    clusters = [[] for _ in range(k)]

    # Gán từng điểm vào cụm gần nhất
    for point_index, point in enumerate(data):
        distances = []

        for centroid in centroids:
            distance = euclidean_distance(point, centroid)
            distances.append(distance)

        # Lấy ra khoảng cách gần nhất với index tương ứng với 3 cụm
        nearest_cluster = distances.index(min(distances))
        clusters[nearest_cluster].append(point)

        print(
            f"P{point_index + 1} = {point} | "
            f"d(C1) = {distances[0]:.3f}, "
            f"d(C2) = {distances[1]:.3f}, "
            f"d(C3) = {distances[2]:.3f} "
            f"=> Cụm {nearest_cluster + 1}"
        )

    # Tính tâm cụm mới
    new_centroids = []

    for i in range(k):
        new_centroid = calculate_centroid(clusters[i])

        # Nếu cụm rỗng thì giữ nguyên tâm cũ
        if new_centroid is None:
            new_centroid = centroids[i]

        new_centroids.append(new_centroid)

    print("\nCác cụm:")

    for i, cluster in enumerate(clusters, start=1):
        print(f"Cụm {i}: {cluster}")

    print("\nTâm cụm mới:")

    for i, centroid in enumerate(new_centroids, start=1):
        print(f"C{i}_new = {centroid}")

    # Kiểm tra hội tụ
    if centroids_equal(centroids, new_centroids):
        print("\nTâm cụm không thay đổi.")
        print("Thuật toán đã hội tụ.")
        break

    centroids = new_centroids
```
